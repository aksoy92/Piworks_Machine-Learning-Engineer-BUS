```r
df <-as.data.frame(read_csv("municipality_bus_utilization.csv"))
df
```

```r
df$date <- as.Date(df$timestamp)
df$time <- format(df$timestamp,"%H:%M:%S")
df$hour = format(as.POSIXct(df$time,format="%H:%M:%S"),"%H")
df$hour<-as.numeric(df$hour)


##hour 7 causes NA values in rows. We consider it as 8 am

df[df$hour==7,]$hour <-8
```

```r
install.packages("visdat")
library("visdat")
vis_miss(df)

```

```r

###

# 
# Df for index to catch NA values
# 

## date dataframe contains all day values.

datedf<-data.frame(matrix(ncol=3,nrow=77))
colnames(datedf)<-c("date","municipality_id","hour")

listdatedf<-data.frame(seq(as.Date("2017-06-04"), as.Date("2017-08-19"), by="days"))

datedf[,1]<- listdatedf[,1]

emptydf<-data.frame(matrix(ncol=3,nrow=1))
colnames(emptydf)<-colnames(datedf)
emptydf$date<-as.Date(emptydf$date, format= "%Y.%m.%d")

for (i in 0:9){
  for(x in 8:16){
    datedf$municipality_id<-i
    datedf$hour<-x
    emptydf<-rbind(emptydf,datedf)
  }
}

alldaydf<- data.frame (emptydf %>% arrange(emptydf$date))
alldaydf<-na.omit(alldaydf)

#### up to here, we finds na values( no usage values for corresponding days
# Therefore, we have na values in merged_df (%2 missing))

merged_df<-merge(x=alldaydf,y=df,by=c("date","municipality_id","hour"),all.x=TRUE)

####  merged_df (%4 missing))
vis_miss(merged_df)

```

```r
#### NA VALUES 
namerged_df<- merged_df[is.na(merged_df$usage),]
namerged_df
```

```r

df<-merged_df


#create index variable for each day (Date)
df$dateindex <- c(1,1+cumsum(diff(df$date)!=0))
df <- df %>% arrange(municipality_id,dateindex,time)


# SELECT MAX USAgE VALUE PER hour
df<-as.data.frame(df %>% group_by(dateindex,hour,municipality_id) %>% top_n(1, usage))

df
```

```r

#drop minutes
df$minute<-NULL


# adding features 

#dayofweek
df$weekday = as.numeric(format(df$date, format = "%u"))

#dayofyear
df$dayofyear<- lubridate::yday(df$date)



#weekofyear
df$weekofyear<-lubridate:: week(df$date)


#dayofmonth
df$dayofmonth<- lubridate::mday(df$date)

#quarter 

df$quarter <- quarters(df$date)
df[df$quarter == "Q2",]$quarter <-2 
df[df$quarter == "Q3",]$quarter <-3



#month #day #year

df<-df %>%
  dplyr::mutate(year = lubridate::year(date), 
                month = lubridate::month(date), 
                day = lubridate::day(date))







# weekend & weekdays

df$weekend<-0
df[df$weekday==6|df$weekday==7,]$weekend <-1

df$workingdays<-1
df[df$weekend==1,]$workingdays <-0


# morning noon after noon

df$morning<-0
df$noon<-0
df$afternoon<-0

df[df$hour<12,]$morning<-1
df[df$hour==12,]$noon<-1
df[df$hour>12,]$afternoon<-1


#Hourofyear 

## function
yhour <- function(time) {(lubridate::yday(time) - 1) * 24 +lubridate:: hour(time)}


df$hourofyear<-yhour(df$timestamp)

df


```

```r


#hotcoding

#municipaly

df$munic0<-0
df$munic1<-0
df$munic2<-0
df$munic3<-0
df$munic4<-0
df$munic5<-0
df$munic6<-0
df$munic7<-0
df$munic8<-0
df$munic9<-0



df<- df %>% mutate(munic0 = ifelse(df$municipality_id == 0, 1, 0),
                   munic1 = ifelse(df$municipality_id == 1, 1, 0),
                   munic2 = ifelse(df$municipality_id == 2, 1, 0),
                   munic3 = ifelse(df$municipality_id == 3, 1, 0),
                   munic4 = ifelse(df$municipality_id == 4, 1, 0),
                   munic5 = ifelse(df$municipality_id == 5, 1, 0),
                   munic6 = ifelse(df$municipality_id == 6, 1, 0),
                   munic7 = ifelse(df$municipality_id == 7, 1, 0),
                   munic8 = ifelse(df$municipality_id == 8, 1, 0),
                   munic9 = ifelse(df$municipality_id == 9, 1, 0))


##Quarters Hotcoding

df$Q2<-0
df$Q3<-0

df<- df %>% mutate(
  Q2 = ifelse(df$quarter == 2, 1, 0),
  Q3 = ifelse(df$quarter == 3, 1, 0),
)

df
```

```r
df$municipality_id <- NULL
df$time<-NULL
df$timestamp<-NULL

```

```r
##0 (NA) values examination
## first case drop na values

df<- na.omit(df)
```

```r

# TEST TRAIN

train <- df[df$date<'2017-08-05',]
test <- df[df$date>'2017-08-05',]

#drop irrelevant columns from train and test set
train$date<- NULL
train$dateindex<- NULL
train$quarter<- NULL
train$year<- NULL

test$date<- NULL
test$dateindex<- NULL
test$quarter<- NULL
test$year<- NULL







train_x <- train %>% dplyr::select(-usage)
train_y <- train$usage

test_x <- test %>% dplyr::select(-usage)
test_y <- test$usage




# Modellerin Egitilmesi

ml_formula <- formula(usage ~ total_capacity+ hour+ weekday+dayofyear 
                      + weekofyear  + dayofmonth 
                      + Q2 + Q3  + month + day 
                      + weekend + workingdays+morning
                      + noon + afternoon + munic0 
                      + munic1 + munic2+munic3+munic4
                      + munic5 + munic6 +munic7 +munic8 
                      + munic9  )



###Linear Regression

ml_linear<-  lm(ml_formula,train)



## Random Forest
ml_rf <- randomForest::randomForest(train_x, train_y, importance = TRUE)


## Gradient Boosted Tree
ml_gbt<-gbm(ml_formula, data = train ,
            distribution = "gaussian", 
            n.trees = 5000,
            interaction.depth = 1,
            shrinkage = 0.01,
            cv.folds = 5)


#ML models' List

ml_models <- list(
  "Lineer Reg" = ml_linear,
  "Random Forest" = ml_rf,
  "Gradient Boosted Trees" = ml_gbt
)

# Skorlama islemi icin fonksiyon yazilmasi
score_test_data <- function(model, data = test){
  
  test_x <- data %>% dplyr::select(-usage)
  test_y <- data$usage
  result_df<-data.frame (
    defaultSummary(data.frame(obs = test_y,
                              pred = predict(model, test_x)))
  )
  summary_df<-as.data.frame(matrix(ncol = 3, nrow = 1))
  colnames(summary_df)<-rownames(result_df)
  summary_df[1:3]<-result_df[1:3,1]
  
  summary_df
}


ml_score <- lapply(ml_models, score_test_data)


```
